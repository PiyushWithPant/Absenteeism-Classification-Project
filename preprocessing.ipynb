We need to give numeric meaning to our categorical numbers in this column so we will use DUMMY VARIABLES

```python
# Dummy variable = 1 if present, 0 if not

reason_column = pd.get_dummies(
    rawData['Reason for Absence'],
    
    # we wanna remove the first reason i.e REASON 0 to avoid Multicollinearity
    
    drop_first= True

)

reason_column
```

NO MULTICOLLINEARTY = to apply Linear Regression

if we merge Dummy variable in the in the main data, 
DUMMY VAR and REASON FOR ABSENSE would convey same meaning to the model, 
hence it they will make it multicollinear, so we need to drop the latter

```python


rawData = rawData.drop(
    ['Reason for Absence'],
    axis = 1
)
```

If we add the dummy variable again now, we will have over 40 FEATURES, which is a lot for 700 dataset, so we need to group the data

To class/group the REASONS FOR ABSENSE, you need the table which contains reasons and their respective numbers, after that you need to understand each and group them like

Group A = all diseases like [1 to 14]

Group B = pregnancy [15 to 17]

Group C = Poisoning of different kinds [18 - 21]

Group D = Light reasons [22-28]

```python
# reason_group_a = reason_column.loc[ : , '1':'14']

# above code will give us the dataset from 1 to 14, but we want only single summarized column,
# so we will sum them to get one value which was responsible for leave, so we will use MAX
```

```python
reason_group_a = reason_column.loc[ : , 1:14].max(axis=1) # : means all rows
reason_group_b = reason_column.loc[ : , 15:17].max(axis=1)
reason_group_c = reason_column.loc[ : , 18:21].max(axis=1)
reason_group_d = reason_column.loc[ : , 21:].max(axis=1)
```

```python
reason_group_a
```

#### Concatenating the preprocessed feature

```python
rawData = pd.concat(
    [rawData, reason_group_a, reason_group_b, reason_group_c, reason_group_d],
    axis= 1
)

rawData
```

Since we have to use many types of preprocessing techniques so we use CHECKPOINTS to ease our work. It is nothing but saving our work at a particular point

```python
# CHECKPOINT 1 - REASON FOR ABSENSE PREPROCESSED

data1 = rawData.copy()

data1
```

DATE column analysis

```python
type(data1['Date'])
```

```python
type(data1['Date'][0])
```

```python
data1['Date'] = pd.to_datetime(
    data1['Date'],
    format = "%d/%m/%Y"
    )

data1.info()

# Now the type is Datetime and not a string
```

Extracting the month value from date

```python
data1['Date'][0]
```

```python
data1['Date'][0].month
```

```python
list_months = []

for i in range(data1.shape[0]):    # since there are 700 data rows
    list_months.append(data1['Date'][i].month)
    
    
list_months
```

Adding Month value as a column in dataframe

```python
data1['Month Value'] = list_months

data1.info()
```

Extract day of the week

```python
data1['Date'][11].weekday()
```

```python
def date_to_weekday(date):
    return date.weekday()
```

using APPLY() to apply our function to each row value of Date

```python
data1['Day of the Week'] = data1['Date'].apply(date_to_weekday)

data1.head()
```

```python
# Removing date

data1 = data1.drop(['Date'], axis = 1)
```

Transforming the education column

```python
data1['Education'].unique()
```

1 = High School only

2 = Graduate

3 = Post Graduate

4 = PHD/Scientist

```python
data1['Education'].value_counts()
```

```python
# Since 2,3,4 are very low, so we will make them into 1 category

data1['Education'] = data1['Education'].map({
    1 : 0,
    2 : 1,
    3 : 1,
    4 : 1,
    
    # 0 = NO high education (Not graduate)
    # 1 = High education (Graduate)
})

data1.head()
```
